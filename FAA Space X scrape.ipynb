```python
import requests
from bs4 import BeautifulSoup
import re
import unicodedata
import pandas as pd
import numpy as np
```

```python
permit_url = "https://www.faa.gov/data_research/commercial_space_data/launches/?type=Permitted"
license_url = "https://www.faa.gov/data_research/commercial_space_data/launches/"
```

```python
permit_data = requests.get(permit_url)
permit_data.status_code
```

```python
license_data = requests.get(license_url)
license_data.status_code
```

```python
permit_soup = BeautifulSoup(permit_data.text)
license_soup = BeautifulSoup(license_data.text)
```

```python
permit_table = permit_soup.find_all('table')
license_table = license_soup.find_all('table')
```

```python
#get table headers
table_headers = license_soup.find_all('th')
```

```python
#append headers to a list for Df creation later
table_header_list = []

for element in table_headers:
    print(element.text)
    table_header_list.append(element.text)
```

```python
table_header_list
```

```python
#shave down just rows in table
permit_rows = permit_soup.find_all('tr')
license_rows = license_soup.find_all('tr')
```

```python
permit_row_data = []
temporary_list = []
for row in permit_rows:
    tds = row.find_all('td')
    for td in tds:
        
        temporary_list.append(td.text)
    if 'Space Exploration Technologies Corporation' in temporary_list:
        permit_row_data.append(temporary_list)
        temporary_list = []
    else:
        temporary_list = []

license_row_data = []
temporary_list = []
for row in license_rows:
    tds = row.find_all('td')
    for td in tds:
        
        temporary_list.append(td.text)
    if 'Space Exploration Technologies Corporation' in temporary_list:
        license_row_data.append(temporary_list)
        temporary_list = []
    else:
        temporary_list = []
```

```python
final_list = []
temp_list = []
for row in permit_row_data:
    for i, element in enumerate(row):
        if i == 0:
            date = element.split()[0]
            date = date[:4]+'-'+date[4:6]+'-'+date[6:8]
            temp_list.append(date)
            continue
        if i == 1:
            element = re.sub("\s+", "", element)
            temp_list.append(element)
            continue
        temp_list.append(element)
            
    final_list.append(temp_list)
    temp_list = []
        
print(final_list)


```

```python
last_list =[]
temp_list = []
for row in license_row_data:
    for i, element in enumerate(row):
        if i == 0:
            date = element.split()[0]
            date = date[:4]+'-'+date[4:6]+'-'+date[6:8]
            temp_list.append(date)
            continue
        if i == 1:
            element = re.sub("\s+", "", element)
            temp_list.append(element)
            continue
        temp_list.append(element)
        
    final_list.append(temp_list)  
    temp_list = []
        
print(final_list)
```

```python
last_list = []
for item in final_list:
    
    if 'Falcon 9' in item or 'Falcon 9-R' in item:

        last_list.append(item)
last_list
```

```python
df = pd.DataFrame(last_list, columns = table_header_list)
```
