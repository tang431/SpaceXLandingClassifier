```python
#average the missing values for payload mass and PayloadMass
# Calculate the mean value of PayloadMass column
mean=combined_df['PayloadMass'].mean()
# Replace the np.nan values with its mean value
combined_df['PayloadMass'].replace(np.nan,mean, inplace=True)
combined_df['PayloadMass']= combined_df['PayloadMass'].astype(int)
```

```python
combined_df['Payload mass'].unique()
```

```python
def parse_payload_mass(x):
    try:
        
        split_string = x.split()[0]
        if len(split_string) ==1:
            return np.nan
        else:
            number = ''.join(char for char in split_string if char.isnumeric())
            if len(number) > 5:
                return 5500 #because there was one instance of a range, and return the average of the range
            else:
                int_number = int(number)
                return int_number
        
    except:
        
        return np.nan
```

```python
combined_df['Payload mass'] = combined_df['Payload mass'].apply(lambda x: parse_payload_mass(x))
```
